```python
import sqlite3
import pandas as pd
```

```python
# Create a connection to the SQLite database
conn = sqlite3.connect('flight.db')

# Execute a query and read the results into a DataFrame
df = pd.read_sql_query("SELECT * FROM Flights", conn)

df
```

```python
# read in the reference dataframe
ref_df = pd.read_csv('archive/airports.csv')

# define the column to use as the reference
ref_col = 'origin'

# loop through each row in the original dataframe
for index, row in df.iterrows():
    # get the value of the reference column for the current row
    ref_val = row[ref_col]
    # use the reference value to get the corresponding data from the reference dataframe
    latitude_deg = ref_df.loc[ref_df['ident'] == ref_val, 'latitude_deg'].values[0]
    longitude_deg = ref_df.loc[ref_df['ident'] == ref_val, 'longitude_deg'].values[0]
    # add the new column values to the original dataframe
    df.at[index, 'latitude_deg'] = latitude_deg
    df.at[index, 'longitude_deg'] = longitude_deg

df
```

```python
# Convert the DataFrame to a CSV file
df.to_csv('filghts-and-airports.csv', index=True)
```
